```python
import matplotlib.pyplot as plt
import numpy as np
```

<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 200)
tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
# Membuat figure dan axes dengan 3 kolom dan 1 baris
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18,8), sharey=True)

# Membuat subplot pertama
ax[0].plot(x, tan, color='b', label='tanX')
ax[0].set_title('Grafik tanx')

# Membuat subplot kedua
ax[1].plot(x, sin, color='r', label='sinX')
ax[1].set_title('Grafik sinx')

# Membuat subplot ketiga
ax[2].plot(x, cos, color='g', label='cosX')
ax[2].set_title('Grafik cosx')

# Membuat legend dan menampilkan hasil plot
fig.legend(loc='upper center')
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
# code here
# Membuat axes utama
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(x, y)
ax.set_title('Grafik CosX')

# Membuat inset axes pojok kiri bawah
ins2 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins2.plot(x, y, color='b')
ins2.plot(x, y2, color='r')
ins2.set_title('Grafik CosX^2')

# Membuat inset axes pojok kanan bawah
ins3 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins3.plot(x, y, color='b')
ins3.plot(x, y3, color='g')
ins3.set_title('Grafik CosX^3')

# Membuat inset axes pojok kanan atas
ins4 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins4.plot(x, y, color='b')
ins4.plot(x, y4, color='k')
ins4.set_title('Grafik CosX^4')

# Membuat inset axes pojok kiri atas
ins5 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins5.plot(x, y, color='b')
ins5.plot(x, y5, color='y')
ins5.set_title('Grafik CosX^5')

plt.show()
```
