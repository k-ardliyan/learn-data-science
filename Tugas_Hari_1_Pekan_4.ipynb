1. Machine learning adalah bidang ilmu yang mengajarkan bagaimana caranya 
mengajari computer/machine untuk belajar dari data yang ada kemudian membuat prediksi terhadap data baru.

2. Feature (atribut) adalah data yang digunakan sebagai input untuk prediksi dalam machine learning, sedangkan target (label) adalah data hasil prediksi yang didapatkan setelah proses training data

3. Supervised learning adalah jenis machine learning yang labelnya sudah diketahui, sedangkan unsupervised learning adalah jenis machine learning yang labelnya belum diketahui atau model tersebut yang menentukan labelnya sendiri
Supervised learning terbagi menjadi 2, yaitu klasifikasi dan regresi. Klasifikasi memiliki output yang diskrit atau hanya membedakan data yang ada, sedangkan regresi memiliki output yang continuous atau berupa nilai dari sesuatu dalam angka.

4. Hyperparameter merupakan angka konfigurasi yang didefinisikan di awal sebagai bagian dari input dan tidak berkaitan dengan model machine learning yang digunakan, sedangkan parameter merupakan data yang dihasilkan dan bisa diestimasi dari data dan berkaitan dengan model yang digunakan.

5. Terdapat 5 langkah dalam mengaplikasikan algoritma dalam machine learning, antara lain :

  * Memisahkan antara data feature dengan target (label)
  * Melakukan training pada model dengan method .fit()
  * Mengaplikasikan model ke dalam data test dengan method .predict()

<h1>Soal 2: Pemahaman Machine Learning</h1>

Pelajarilah secara garis besar suatu model/algoritma machine learning, kemudian aplikasikan untuk membuat model bagi data (variable x, y) di bawah ini. Kemudian buatlah prediksi terhadap data training dan data baru di interval 20-30.

- Plot data asli, data hasil prediksi terhadap data training, data hasil prediksi terhadap data baru
- Tunjukan beberapa parameter yang dimiliki model yang telah kalian buat

```python
import matplotlib.pyplot as plt
import numpy as np

rng = np.random.RandomState(42)
x = 20 * rng.rand(50)
y = x**2 + 2 * x + - 1 + rng.randn(50)

# data yang akan di prediksi
x_new = np.arange(20, 30, 0.5)

fig, ax = plt.subplots(figsize=(12, 6))
ax.scatter(x, y)
ax.set_xlabel('X-Value')
ax.set_ylabel('Y-Value')
plt.show()
```

```python
# Code here
# Melakukan reshape pada data train dan data test
x = x.reshape(-1, 1)
x_new = x_new.reshape(-1, 1)
```

```python
# Import DecisionTreeRegressor dan plot_tree
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree

# Membuat model DecisionTreeRegressor
model = DecisionTreeRegressor(max_depth=6, min_samples_leaf= 0.1)

# fit data ke dalam model
model.fit(x, y)

# Melakukan prediksi terhadap data train
y_train = model.predict(x)

# prediksi terhadap data test
y_test = model.predict(x_new)
```

```python
# Membuat plot data asli, data train, dan data test
fig, ax = plt.subplots(figsize=(12,8))
ax.scatter(x, y, color='b', label='Data Asli')
ax.scatter(x, y_train, color='y', label='Prediksi terhadap data training')
ax.plot(x_new, y_test, color='g', label='Prediksi terhadap data baru')
plt.legend()
plt.show()
```

```python
print(model.feature_importances_)
print(model.n_features_)
```

```python
# Membuat tree plot dari model
fig, ax = plt.subplots(figsize=(12, 8))
plot_tree(model)
plt.show()
```
